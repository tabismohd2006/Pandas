creating series as String

```python
import pandas as pd
names = pd.Series(["Aman", "Ravi", "aryan", "nadim"])
print(names)
```

creating a series with custom index

```python
import pandas as pd
marks = pd.Series(
    [85,90,78],
    index = ["math","science", "hindi"]
)
print(marks)
```

create a series from dictonary

```python
import pandas as pd
student = pd.Series({
    "Name": "Aman",
    "Age": 20,
    "City": "Delhi"
})
print(student)
```
