module 5 missing Value in pandas

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Riya", "Karan", "Priya", np.nan],
    "Age": [22, np.nan, 25, 28, 24],
    "City": ["Delhi", "Mumbai", np.nan, "Lucknow", "Delhi"],
    "Salary": [50000, 60000, np.nan, 75000, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)
print(df.isnull())
print(df.isnull().sum())
print(df.notnull())
filled = df.fillna("Unknown")
print(filled)
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)
df.fillna({
    "Name": "Not Available",
    "City": "Unknown",
    "Salary": 0
}, inplace=True)

print(df)

df.dropna()
```

```python
import pandas as pd
data = {
    "ID":[101, 102, 103, 103, 104],
    "Name": ["Aman", "Riya", "Rahul", "Rahul", "Karan"],
    "city": ["Delhi", "Mumbai", "Bangalore", "Bangalore", "Lucknow"]
}
df = pd.DataFrame(data)
print("original DataFrame")
print(df)
df =df.drop_duplicates()
print(df)
```

```python
import pandas as pd
data = {
    "ID":[101, 102, 103,],
    "salary": [45000, 60000, 55000],

}
df = pd.DataFrame(data)
print("Before Conversion")
print(df.dtypes)
df["ID"] = df["ID"].astype(int)
df["salary"] = df["salary"].astype(float)
print("\nAfter Conversion")
print(df.dtypes)
print(df)
```

lower() upper(), contains(), split() replace(), strip()

```python
import pandas as pd
df= pd.DataFrame({
    "Name": ["Aman", "Riya", "Rahul", "Priya", "Karan"],

})
print(df)
df["Name"] = df["Name"].str.lower()
print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Amit", "Priya", "Rahul", "Neha", "David"]
})

print(df)

df["Lower"] = df["Name"].str.lower()

print(df)

df["Upper"] = df["Name"].str.upper()

print(df)

df["Contains_a"] = df["Name"].str.contains("a", case=False)

print(df)


df["Split"] = df["Name"].str.split()

print(df)


df["Replaced"] = df["Name"].str.replace("a", "A")

print(df)

df["Stripped"] = df["Name"].str.strip()

print(df)
```
